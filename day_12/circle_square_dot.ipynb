```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000000
counter_circle=0
x=np.random.uniform(-1,1,N)
y=np.random.uniform(-1,1,N)

for i in range(N):
    if np.sqrt(x[i]**2 + y[i]**2)<=1:
        counter_circle+=1
print(counter_circle)
pi=4*counter_circle/N
print("pi:",pi)
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000
inside = []
counter_circle=0
x=np.random.uniform(-1,1,N)
y=np.random.uniform(-1,1,N)

for i in range(N):
    if np.sqrt(x[i]**2 + y[i]**2) < 1:
        inside.append((x, y))
        
plt.figure(figsize=(6,6))
plt.scatter([x[0] for x in inside], [y[1] for y in inside], marker=".", alpha=0.5)

print("Value of pi : ", 4 * len(inside)/float(N))
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000000
counter_circle=0
x=np.random.uniform(-1,1,N)
y=np.random.uniform(-1,1,N)

for i in range(N):
    if np.sqrt(x[i]**2 + y[i]**2)<=1:
        counter_circle+=1
print(counter_circle)
pi=4*counter_circle/N
print("pi:",pi)


circle = plt.Circle((0,0), 1, color='black')
fig, ax = plt.subplots()
ax.add_patch(circle)
ax.scatter(x,y)
ax.set_aspect("equal")
plt.xlim([-1.1,1.1])
plt.ylim([-1.1, 1.1])
plt.xlabel("X")
plt.ylabel("Y")
```
