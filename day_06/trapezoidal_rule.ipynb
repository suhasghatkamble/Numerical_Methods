```python
import numpy as np
def IT(f, a, b, n):
    h = (b - a) / n
    sum = 0
    for i in range(int(n)):
        t = (f(a + i*h) + f(a + (i + 1)*h))
        sum = sum + t
    return sum * h / 2
def f(x):
    y = 2 + 2 * x + x**2 + np.sin(2*np.pi*x) + np.cos(2*np.pi*x/0.5)
    return y
print(IT(f, 0 , 1.5, 1.0))
print(IT(f, 0, 1.5, 9.0))
```

```python
import numpy as np
import scipy.integrate
def IT(f, a, b, n):
    h = (b - a) / n
    sum = 0
    for i in range(int(n)):
        t = (f(a + i*h) + f(a + (i + 1)*h))
        sum = sum + t
    return sum * h / 2

def f(x):
    y = 2 + 2 * x + x**2 + np.sin(2*np.pi*x) + np.cos(2*np.pi*x/0.5)
    return y
print(IT(f, 0 , 1.5, 1.0))
print(IT(f, 0, 1.5, 9.0))
print(scipy.integrate.quad(f, 0, 1.5)[0])
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 6, 100)
y = -x**2 + 4*x + 5
y1 = -2*x + 4
y2 = -2
plt.plot(x, y, label='f(x)')
plt.plot(x, y1, label='f\'(x)')
plt.plot(x, y2, label='f"(x)')

plt.legend()
plt.grid(True)
plt.show()
```
