```python
#Slope intercept form

import matplotlib.pyplot as plt

x_axis=[]
y_axis=[]
m=2
b=3

for x in range(0,100):
    x_axis.append(x)
    y_axis.append(b-(m*x)) #3-(2*1)

plt.plot(x_axis,y_axis) 
plt.xlabel('x')
plt.ylabel('y')
plt.show()

```

```python
#Point-slope form
# y-y1=m(x-x1)  , y-3=6(x-2)

import matplotlib.pyplot as plt

x_axis=[]
y_axis=[]
m=6
y1=3
x1=2

for x in range(-50,50):
    x_axis.append(x)
    y_axis.append(m * (x - x1) + y1)

plt.plot(x_axis,y_axis) 
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
