```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
p1 = np.random.normal(5,1,500000)
p2 = np.random.normal(10,1,500000)
p3 = np.random.normal(15,1,500000)

y = p1 + p2 + p3


count_34 = np.count_nonzero(y>=34)


probability = count_34/len(y)
count_34, probability, probability * 100
```

```python
count, bin, ignored  = plt.hist(y, bins=30, density=True, color='skyblue')
plt.vlines(34,count.min(), count.max())
plt.plot()
```
